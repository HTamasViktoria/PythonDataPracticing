```python
#országonként milyen típust készítettek legtöbbet?
df_main_type_by_country = df.groupby(['country', 'type']).size().reset_index(name='count')

df_main_type_by_country_sorted = df_main_type_by_country.sort_values(by='count', ascending=False)

df_pivot = df_main_type_by_country.pivot_table(index='country', columns='type', values='count', fill_value=0)

# Összes típusos oszlop hozzáadása
df_pivot['total'] = df_pivot.sum(axis=1)

# Rendezés csökkenő sorrendbe az összes oszlop alapján
df_pivot_sorted = df_pivot.sort_values(by='total', ascending=False)
df_pivot_sorted
```

```python
df_exploded_cast = df[~df['cast'].isna()]
df_exploded_cast['cast'] = df['cast'].str.split(',')
df_exploded_cast = df_exploded_cast.explode('cast', ignore_index=True)
df_exploded_cast['cast'].value_counts().head(30)
```

```python
#melyik listed-in-ből van a legtöbb
df_exploded_categories = df.copy()
df_exploded_categories['listed_in'] = df['listed_in'].str.split(',')
df_exploded_categories= df_exploded_categories.explode('listed_in', ignore_index=True)
#df_exploded_categories.sample(10, random_state=42)
category_counts = df_exploded_categories['listed_in'].value_counts().head(5)
plt.bar(category_counts.index, category_counts)
plt.xticks(rotation=45, ha='right')
```

```python
#mennyi most a normál filmek hosszának az átlaga?

df_movies = df[~(df['type'] == 'TV Show')]
# Ha van NaN, akkor először tisztítjuk a NaN-t, hogy a replace működjön
df_movies['duration'] = df_movies['duration'].fillna('0 min')  # NaN értékek helyettesítése egy alapértelmezett értékkel
df_movies['duration'] = df_movies['duration'].str.replace(' min', '')  # A 'min' eltávolítása
df_movies['duration'] = pd.to_numeric(df_movies['duration'], errors='coerce')  # Konvertálás szám típusra, a nem konvertálható értékek 'NaN' lesznek

df_movies = df_movies.dropna(subset=['duration'])
df_movies['duration'].median()



```

```python
#típusonként az átlagos hossz
df_movies['listed_in'] = df_movies['listed_in'].str.split(',')
df_movies= df_movies.explode('listed_in', ignore_index=True)
# Csoportosítás 'listed_in' alapján, majd medián számítása a 'duration' oszlopra
df_movies.groupby('listed_in')['duration'].agg('median')
```
